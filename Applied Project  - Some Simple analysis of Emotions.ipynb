```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame
def plot_distributions(df):
    # Loop through each column
    for col in df.columns:
        plt.figure(figsize=(8, 4))
        if df[col].dtype in ['int64', 'float64']:
            # For numerical columns, use a histogram
            sns.histplot(df[col], kde=True)
            plt.title(f'Distribution of {col}')
        else:
            # For categorical columns, use a countplot
            sns.countplot(y=df[col])
            plt.title(f'Count plot of {col}')
        plt.tight_layout()
        plt.show()

# Example usage:
# df = pd.read_csv('your_file.csv')
plot_distributions(emotions)
```

```python
import pandas as pd

# Assuming df is your dataset
def calculate_summary_statistics(df):
    summary = df.describe().T  # Transpose the output to have columns as rows
    summary['missing'] = df.isnull().sum()  # Add missing values count
    summary['unique'] = df.nunique()  # Add number of unique values for each column
    summary['dtype'] = df.dtypes  # Add data type for each column
    
    return summary

# Example usage:
# df = pd.read_csv('your_dataset.csv')  # Replace with your dataset loading method
summary_df = calculate_summary_statistics(emotions)

# Display the aggregated summary statistics
print(summary_df)
```
